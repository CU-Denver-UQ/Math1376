```python
from bs4 import BeautifulSoup
import urllib

url = 'http://www.denverbroncos.com/team/roster.html'

r = urllib.request.urlopen(url).read()

print(r)
```

```python
# soup = BeautifulSoup(r, 'lxml')
soup = BeautifulSoup(r, 'html.parser')
print(soup.prettify())
```

```python
broncos_main = soup.find(id="main-content")
print(broncos_main.prettify())
```

```python
tr_tags = broncos_main.find_all("tr", {"class":[" loop-first loop-odd ", 
                                          " loop-odd loop-last ",
                                          " loop-even loop-last ",                                        
                                          " loop-odd ", " loop-even "]})
print(tr_tags)
```

```python
print(len(tr_tags))
```

```python
print(tr_tags[0])
```

```python
print("{:8s} {:30s} {:5s} {:6s} {:5s} {:3s} {:3s} {:40s}"
      .format('number', 'name', 'pos', 'ht', 'wt', 'age', 'exp', 'college'))                                           
for tr in tr_tags:  
    number = tr.find_all("td", {"class": "col-jersey"})[0].get_text().strip()
    name = tr.find_all("a", {"class": "player-card-tooltip"})[0]["title"]
    pos = tr.find_all("td", {"class": "col-position"})[0].get_text()
    ht = tr.find_all("td", {"class": "col-height"})[0].get_text()
    wt = tr.find_all("td", {"class": "col-weight"})[0].get_text()
    age = tr.find_all("td", {"class": "col-bd"})[0].get_text()
    exp = tr.find_all("td", {"class": "col-exp"})[0].get_text().strip()
    college = tr.find_all("td", {"class": "col-college"})[0].get_text()
    # print(number, name, pos, ht, wt, age, exp, college)
    print("{:8s} {:30s} {:5s} {:6s} {:5s} {:3s} {:3s} {:40s}".format(number, name, pos, ht, 
                                                wt, age, exp, college))
```
